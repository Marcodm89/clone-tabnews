#Biblioteca

```python
import pandas as pd
import plotnine as p9
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
dados_brutos_ipca=pd.read_json('https://apisidra.ibge.gov.br/values/t/1737/n1/all/v/2265/p/all/d/v2265%202?formato=json')
dados_brutos_ipca
```

#Tratamento de dados

```python
(
    dados_brutos_ipca.info()
)
```

```python

dados_tratados_ipca=(
    dados_brutos_ipca
    .rename(columns={"V":"IPCA","D3C":"DATA"})
    .query("IPCA not in['Valor','...']")
    .assign(
            DATA=lambda x: pd.to_datetime(x['DATA'],format='%Y%m'),
            IPCA=lambda x: x['IPCA'].astype(float)
            )
    .filter(['DATA','IPCA'])
    .query("DATA>=@pd.to_datetime('2004-01-01')")
    )
dados_tratados_ipca
```

```python
dados_brutos_desempregos=pd.read_json('https://apisidra.ibge.gov.br/values/t/6381/n1/all/v/4099/p/all/d/v4099%201?formato=json')
dados_brutos_desempregos
```

```python
dados_tratados_desemprego=(
    dados_brutos_desempregos
    .rename(columns={"V":"DESEMPREGO","D3C":"DATA"})
    .query("DESEMPREGO not in['Valor','...']")
    .assign(
            DATA=lambda x: pd.to_datetime(x['DATA'],format='%Y%m'),
            DESEMPREGO=lambda x: x['DESEMPREGO'].astype(float)
            )
    .filter(['DATA','DESEMPREGO'])
    )
dados_tratados_desemprego
```

```python
dados_cruzados=(
    dados_tratados_ipca.set_index('DATA')
    .join(dados_tratados_desemprego
    .set_index('DATA'),how='inner')
    .reset_index()
    )
dados_cruzados
```

#Análise de dados

```python
#Como a inflação se comportou no Brasil?
(
    p9.ggplot(dados_tratados_ipca)+
    p9.aes(x='DATA',y='IPCA')+
    p9.geom_line()
)
```

```python
#Qual o périodo com menores e maiores taxas de inflação?
dados_tratados_ipca.query("IPCA==IPCA.min() or IPCA==IPCA.max()")
```

```python
#Qual o périodo com menores e maiores taxas de inflação?
maximo_ipca=dados_tratados_ipca.sort_values('IPCA').set_index('DATA').iloc[0]
minimo_ipca=dados_tratados_ipca.sort_values('IPCA').set_index('DATA').iloc[-1]

maximo_ipca,minimo_ipca
```

```python
#Qual o valor médio da inflação do Brasil e como é a distribuição de seus valores
dados_tratados_ipca.IPCA.describe()
```

```python
(
    p9.ggplot(dados_tratados_ipca)+
    p9.aes(x='IPCA')+
    p9.geom_histogram()
)
```

```python
# O que afeta a inflação? Qual variável se relaciona?
(
    p9.ggplot(dados_cruzados)+
    p9.aes(x='DESEMPREGO',y='IPCA')+
    p9.geom_point()
)
```

```python
modelo=sm.OLS(
    endog=dados_cruzados['IPCA'],
    exog=dados_cruzados['DESEMPREGO'])
modelo=modelo.fit()
print(modelo.summary())
```

```python
modelo2=smf.ols("IPCA~DESEMPREGO",data=dados_cruzados).fit()
print(modelo2.summary())
```
